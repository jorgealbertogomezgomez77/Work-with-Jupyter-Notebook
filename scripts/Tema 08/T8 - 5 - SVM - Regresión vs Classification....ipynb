# Clasificacion de flores iris...

```python
import pandas as pd
import numpy as np
from sklearn import svm, datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
print(iris)
```

```python
X = iris.data[:, :2]
Y = iris.target
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max - x_min) / 100

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
X_plot = np.c_[xx.ravel(), yy.ravel()]
```

```python
C = 1.0
svc = svm.SVC(kernel = "linear", C = C, decision_function_shape = "ovr").fit(X, Y)
Ypred = svc.predict(X_plot)
Ypred = Ypred.reshape(xx.shape)
```

```python
plt.figure(figsize = (16, 9))
plt.contourf(xx, yy, Ypred, cmap = plt.cm.tab10, alpha = 0.3)
plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = plt.cm.tab10)
plt.xlabel("Longitud de los ptalos")
plt.ylabel("Anchura de los ptalos")
plt.xlim(xx.min(), xx.max())
plt.title("SVC par las flores de Iris con Kernel Lineal")
```

```python
C = 1.0
svc = svm.SVC(kernel = "rbf", C = C, decision_function_shape = "ovr").fit(X, Y)
Ypred = svc.predict(X_plot)
Ypred = Ypred.reshape(xx.shape)
```

```python
plt.figure(figsize = (16, 9))
plt.contourf(xx, yy, Ypred, cmap = plt.cm.tab10, alpha = 0.3)
plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = plt.cm.tab10)
plt.xlabel("Longitud de los ptalos")
plt.ylabel("Anchura de los ptalos")
plt.xlim(xx.min(), xx.max())
plt.title("SVC par las flores de Iris con Kernel Radial")
```

```python
C = 1.0
svc = svm.SVC(kernel = "poly", C = C, decision_function_shape = "ovr").fit(X, Y)
Ypred = svc.predict(X_plot)
Ypred = Ypred.reshape(xx.shape)
plt.figure(figsize = (16, 9))
plt.contourf(xx, yy, Ypred, cmap = plt.cm.tab10, alpha = 0.3)
plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = plt.cm.tab10)
plt.xlabel("Longitud de los ptalos")
plt.ylabel("Anchura de los ptalos")
plt.xlim(xx.min(), xx.max())
plt.title("SVC par las flores de Iris con Kernel Radial")
```

```python
C = 1.0
svc = svm.SVC(kernel = "sigmoid", C = C, decision_function_shape = "ovr").fit(X, Y)
Ypred = svc.predict(X_plot)
Ypred = Ypred.reshape(xx.shape)
plt.figure(figsize = (16, 9))
plt.contourf(xx, yy, Ypred, cmap = plt.cm.tab10, alpha = 0.3)
plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = plt.cm.tab10)
plt.xlabel("Longitud de los ptalos")
plt.ylabel("Anchura de los ptalos")
plt.xlim(xx.min(), xx.max())
plt.title("SVC par las flores de Iris con Kernel Radial")
```

# Resumen final de la clasificacin Iris...

```python
def svm_iris(C = 1.0, gamma = 0.01, kernel = "rbf"):
    import pandas as pd
    import numpy as np
    from sklearn import svm, datasets
    import matplotlib.pyplot as plt
    %matplotlib inline
    
    iris = datasets.load_iris()
    
    X = iris.data[:, :2]
    Y = iris.target
    
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    h = (x_max - x_min) / 100

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    X_plot = np.c_[xx.ravel(), yy.ravel()]

    svc = svm.SVC(kernel = kernel, C = C, gamma = gamma, decision_function_shape = "ovr").fit(X, Y)
    Ypred = svc.predict(X_plot)
    Ypred = Ypred.reshape(xx.shape)
    plt.figure(figsize = (16, 9))
    plt.contourf(xx, yy, Ypred, cmap = plt.cm.tab10, alpha = 0.3)
    plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = plt.cm.tab10)
    plt.xlabel("Longitud de los ptalos")
    plt.ylabel("Anchura de los ptalos")
    plt.xlim(xx.min(), xx.max())
    plt.title("SVC par las flores de Iris con Kernel " + kernel)
```

```python
from ipywidgets import interact, fixed
```
