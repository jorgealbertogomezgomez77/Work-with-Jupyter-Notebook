```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../../datasets/movies/movies.csv", sep = ";")
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p = 1)
dd2 = distance_matrix(data[movies], data[movies], p = 2)
dd10 = distance_matrix(data[movies], data[movies], p = 10)
```

```python
def dm_to_df(dd, col_name):
    import pandas as pd
    return pd.DataFrame(dd, index = col_name, columns = col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

# Clusterin jerárquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward")
Z
```

```python
plt.figure(figsize = (25, 10))
plt.title("Dendograma jerárquico para el Clusterin")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation = 90.0, leaf_font_size = 10.0)
plt.show()
```

```python
ZZ = linkage(data[movies], "average")
ZZ
```

```python
plt.figure(figsize = (25, 10))
plt.title("Dendograma jerárquico para el Clusterin")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(ZZ, leaf_rotation = 0.0, leaf_font_size = 10.0)
plt.show()
```

```python
ZZZ = linkage(data[movies], "complete")
ZZZ
```

```python
plt.figure(figsize = (25, 10))
plt.title("Dendograma jerárquico para el Clusterin")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(ZZZ, leaf_rotation = 0.0, leaf_font_size = 10.0)
plt.show()
```
